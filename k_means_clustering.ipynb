```python
x = pd.get_dummies(df)
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaled_x = scaler.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
```

```python
#help(KMeans)
```

```python
model = KMeans(n_clusters=2)
```

```python
#fit_transform
cluster_lables = model.fit_predict(scaled_x)
```

```python
cluster_lables
```

```python
x['Cluster']= cluster_lables
```

```python
#x
```

```python
x.corr()
```

```python
x.corr()['Cluster'].iloc[:-2]
```

```python
plt.figure(figsize=(12,6))
x.corr()['Cluster'].iloc[:-2].sort_values().plot(kind='bar')
```

```python
ssd  =[]
for k in range(2,10):
    model = KMeans(n_clusters = k)
    model.fit(scaled_x)
    
    ssd.append(model.inertia_)
```

```python
ssd
```

```python
plt.plot(range(2,10),ssd,"o--")
```

```python
pd.Series(ssd)
```

```python
pd.Series(ssd).diff()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
image_as_array = mpimg.imread('palm_trees.jpg')
```

```python
image_as_array
```

```python
image_as_array.shape
```

```python
image_as_array
```

```python
plt.figure(dpi=150)
plt.imshow(image_as_array)
```

```python
#(H,W,C)
```

```python
(h,w,c) = image_as_array.shape
```

```python
c
```

```python
image_as_array2d = image_as_array.reshape(h*w,c)
```

```python
len(image_as_array.shape)
```

```python
len(image_as_array2d.shape)
```

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters = 7)
```

```python
labels = model.fit_predict(image_as_array2d)
```

```python
labels
```

```python
rgb_codes = model.cluster_centers_.round(0).astype(int)
```

```python
rgb_codes
```

```python
quantized_image = np.reshape(rgb_codes[labels],(h,w,c))
```

```python
plt.imshow(quantized_image)
```
